```python
import numpy as np
patient_heart_rate=np.random.randint(60,101,size=(60,24)).astype(float)
missing_value=np.random.rand(60,24)<0.05
patient_heart_rate[missing_value]=np.nan

np.set_printoptions(threshold=np.inf)
print(patient_heart_rate)

```

```python
#Detect abnormal readings outside the normal range

hourly_average=np.nanmean(patient_heart_rate,axis=0)
print("Avearge heartbeat in hours")

np.set_printoptions(threshold=np.inf)
print(hourly_average)
```

```python
#Detect abnormal readings outside the normal range

abnormal_reading=(patient_heart_rate<60)|(patient_heart_rate>90)

np.set_printoptions(threshold=np.inf)
print("Abnoramal reading")
print(abnormal_reading)
```

```python
#Replace missing values using forward fill logic
for i in range(patient_heart_rate.shape[0]):
    for j in range(1,patient_heart_rate.shape[1]):
        if np.isnan(patient_heart_rate[i,j]):
            patient_heart_rate[i,j]=patient_heart_rate[i,j-1]
            
for i in range(patient_heart_rate.shape[0]):
    if np.isnan(patient_heart_rate[i,0]):
        patient_heart_rate[i,0]=70
 
np.set_printoptions(threshold=np.inf)           
print(patient_heart_rate)

if np.isnan(patient_heart_rate).any():
    print("missing values")
else:
    print("no missing values")
```
