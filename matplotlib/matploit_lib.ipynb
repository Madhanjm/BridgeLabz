```python
df["Age"]= [25,30,22,28,35,26,32,29,31,27,34,33,24]
df.head()
```

```python
#scatter plot:direct realtionship
plt.scatter(df['Age'],df['salary'],color='orange')
plt.show()
```

```python
sort_age=df.sort_values(by='Age')
```

```python
#line plot
plt.plot(sort_age['Age'],df['salary'],color='purple',marker='o',linestyle=':',linewidth=2)
plt.grid()
plt.show()
```

```python
#Bar chart:
plt.bar(sort_age['Age'],df['salary'],color='brown',align='center')
plt.show()
```

```python
#Bivariate - Numerical vs categorical
hr_sal=df[df['dept']=='HR']['salary']
hr_sal
```

```python
it_sal=df[df['dept']=='IT']['salary']
it_sal
```

```python
fin_sal=df[df['dept']=='Finance']['salary']
fin_sal
```

```python
#Box plot
plt.boxplot([hr_sal,it_sal,fin_sal],labels=['HR','IT','Finance'])
plt.show()
```

```python
#pie chart:

salary_by_dept=df.groupby('dept')['salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct="%1.2f", shadow=True,explode=(0.1,0.1,0.1))
plt.axis('equal')
plt.show()
```

```python
#Bar Plot

hr_mean=sum(hr_sal)/len(hr_sal)
it_mean=sum(it_sal)/len(it_sal)
fin_mean=sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(['HR','IT','Finance'],[hr_mean,it_mean,fin_mean],color=['red','green','blue'])
plt.grid()
plt.show()
```

```python
df['experience'] = [2,5,1,3,7,2,6,4,5,3,6,5,1]
df.head()
```

```python
#Bubble Plot:
plt.scatter(df["Age"],df["salary"],s=df["experience"]*20,color="blue",edgecolors="black")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("salary")
plt.show()
```

```python
#2 numerical + 1 categorical
plt.scatter(df["Age"],df['salary'],c=df["dept"].map({"HR":"yellow","IT":"red","Finance":"orange"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Dept")
plt.legend()
plt.show()
```

```python
color={"HR":"yellow","IT":"red","Finance":"orange"}

for dept,color in color.items():
    df_dept=df[df['dept']==dept]
    plt.scatter(df_dept["Age"],df_dept['salary'],c=color,label=dept)
    plt.legend()
```

```python
#Object oriented API
```

```python
fig ,axs=plt.subplots(1,3,figsize=(15,5))

#Line Plot:

axs[0].plot(sort_age['Age'],df['salary'],color='purple',marker='*',linestyle=':',linewidth=2,markersize=2)
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

#histogram:

axs[1].hist(df['salary'],bins=5,color='green')
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

#Box plot:
axs[2].boxplot(df['salary'])
axs[2].set_title("Box plot")
axs[2].set_xlabel("Salary")

plt.savefig("Multiple plots.png")
plt.show()
```

```python
#3-d plot
ax=plt.axes(projection='3d')
ax.scatter(df['Age'],df['salary'],df['experience'])
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_zlabel('Experience')
plt.show()
```

```python
data = {
    "x": [1, 2, 3, 4, 5, 6],
    "y": [10, 20, 30, 25, 35, 40],
    "size": [50, 80, 120, 60, 100, 90],        
    "category": ["A", "B", "A", "B", "A", "B"] 
}

df = pd.DataFrame(data)

colors = df["category"].map({"A": "red", "B": "blue"})

plt.scatter(df["x"], df["y"], s=df["size"], c=colors)

plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Scatter Plot with 3 Numerical and 1 Categorical Value")
plt.show()
```
