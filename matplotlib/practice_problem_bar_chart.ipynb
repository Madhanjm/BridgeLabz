```python
import matplotlib.pyplot as plt
import pandas as pd
# 1.â€‹ Write a Python programming to display a bar chart of the popularity of programming 
# Languages.

data={
    'Programming language':['Java','Python','PHP','JavaScript','C#','C++'],
    'Popularity':[22.2, 17.6, 8.8, 8, 7.7, 6.7 ]
}

df=pd.DataFrame(data)
plt.bar(df['Programming language'],df['Popularity'],color='blue',label='Programming Languages')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.legend()

plt.show()
```

```python
# 2.â€‹ Write a Python programming to display a horizontal bar chart of the popularity of 
# programming Languages.

plt.barh(df['Programming language'],df['Popularity'],color='blue',label='Programming Languages')
plt.ylabel("Programming Languages")
plt.xlabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.legend()

plt.show()
```

```python
# 3.â€‹ Write a Python programming to display a bar chart of the popularity of programming 
# Languages. Use uniform color.

plt.bar(df['Programming language'],df['Popularity'],color='#833AB4',label='Programming Languages')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.legend()

plt.show()
```

```python
# 4.Write a Python programming to display a bar chart of the popularity of programming 
# Languages. Use different color for each bar.  

colors = ['#FF5733', '#33FF57', '#3357FF', '#F333FF', '#33FFF5', '#F5FF33']
plt.bar(df['Programming language'],df['Popularity'],color=colors,label='Programming Languages')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.legend()

plt.show()
```

```python
# 5.Write a Python programming to display a bar chart of the popularity of programming 
# Languages. Attach a text label above each bar displaying its popularity (float value).

colors = ['#FF5733', '#33FF57', '#3357FF', '#F333FF', '#33FFF5', '#F5FF33']
bars=plt.bar(df['Programming language'],df['Popularity'],color=colors,label='Programming Languages')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Languages")
plt.legend()

for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f"{height:.1f}",ha="center",va="bottom")

plt.show()
```

```python
# 10.Write a Python program to create bar plot of scores by group and gender. Use multiple X 
# values on the same chart for men and women.

df={
    'Mens':[20, 34, 30, 35, 26],
    'Womens':[25, 32, 30, 35, 29]
}

group=['G1', 'G2', 'G3', 'G4', 'G5']


plt.bar([i-0.2 for i in range(len(df['Mens']))],df['Mens'],width=0.4,color='orange',label='Mens')
plt.bar([i+0.2 for i in range(len(df['Womens']))],df['Womens'],width=0.4,color='hotpink',label='Womens')
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.title("Scores by Group and Gender")
plt.xticks(range(len(group)), group)
plt.legend()

plt.show()
```

```python
#11.â€‹ Write a Python program to create bar plot from a DataFrame.
data={
    'a':[2,4,6,8,10],
    'b':[4,2,4,2,2],
    'c':[8,3,7,6,4],
    'd':[5,4,4,4,3],
    'e':[7,2,7,8,3],
    'f':[6,6,8,6,2]
}

colors = ['#FF5733', '#33FF57', '#3357FF', '#F333FF', '#33FFF5', '#F5FF33']


df=pd.DataFrame(data)
df.plot(kind='bar',color=colors,edgecolor='black',lw=2)

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Bar Plot from DataFrame")

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 12.â€‹ Write a Python program to create bar plots with error bars on the same figure. Sample Date 
# Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824 
# Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645 

data={'mean Velocity':[0.2474, 0.1235, 0.1737, 0.1824],
      'std Deviation':[0.3314, 0.2278, 0.2836, 0.2645]
}

df=pd.DataFrame(data)
positions=range(len(df))
plt.bar(positions,df['mean Velocity'],yerr=df['std Deviation'],color='yellow',edgecolor='black',lw=2,capsize=5)
plt.xlabel("saamples")
plt.ylabel("Mean Velocity")
plt.title("Bar plots with error bars")
plt.xticks(positions, [f"Sample {i+1}" for i in positions])
plt.show()
```

```python
data={'mean Velocity':[0.2474, 0.1235, 0.1737, 0.1824],
      'std Deviation':[0.3314, 0.2278, 0.2836, 0.2645]
}

df=pd.DataFrame(data)
positions=range(len(df))
bars=plt.bar(positions,df['mean Velocity'],yerr=df['std Deviation'],color='yellow',edgecolor='black',lw=2,capsize=5)

for i ,bar in enumerate(bars):
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f"{df['mean Velocity'][i]*1000}",ha='center',va='bottom')
plt.xlabel("samples")
plt.ylabel("Mean Velocity")
plt.title("Bar plots with error bars")
plt.xticks(positions, [f"Sample {i+1}" for i in positions])
plt.show()
```
