```python
#11. Write a Python program to find the number of elements of an array, length of one 
#array element in bytes and total bytes consumed by the elements.  
array=np.array([1.1,2.1,3.1])
size_of_array=array.size
length_in_bytes=array.itemsize
total_bytes_consumed=array.nbytes
print("Size of the array: ",size_of_array)
print("Length of one array element in bytes: ",length_in_bytes)  
print("Total bytes consumed by the elements of the array: ",total_bytes_consumed) 

```

```python
#12. Write a Python program to find common values between two arrays. 
array1=np.array([1,2,3,4,5])
array2=np.array([2,3,8,9])
comman_value=np.intersect1d(array1,array2)
print("Array1 :",array1)
print("Array2 :",array2)
print("Comman Value between Two Array :",comman_value)
```

```python
#13 Write a Python program to find the set difference of two arrays.
array1=np.array([1,2,3,4,5])
array2=np.array([2,3,8,9])
set_differnce=np.setdiff1d(array1,array2)
print("Array1 :",array1)
print("Array2 :",array2)
print("set differnce between Two Array :",set_differnce)
```

```python
#14. Write a Python program to find the set exclusive-or of two arrays.
array1=np.array([1,2,3,4,5])
array2=np.array([2,3,8,9])
set_exclusive=np.setxor1d(array1,array2)
print("Array1 :",array1)
print("Array2 :",array2)
print("set exclusive-or between Two Array :",set_exclusive)
```

```python
#15. Write a Python program compare two arrays using numpy.  
a=np.array([1,2])
b=np.array([4,5])
greater=np.greater(a,b)
print("a>b")
print(greater)
greater_and_equal=np.greater_equal(a,b)
print("a>=b")
print(greater_and_equal)

less=np.less(a,b)
print("a<b")
print(less)
less_and_equal=np.less_equal(a,b)
print("a<=b")
print(less_and_equal)



```
