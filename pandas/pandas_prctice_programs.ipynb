```python
#2. Write a Python program to convert a Panda module Series to Python list and it's type.
series=pd.Series([1,2,3,4,5])
print("Series:")
print(series)

print("Series to list")
list_series=series.to_list()
print(list_series)
print(type(list_series))
```

```python
#3. Write a Python program to add, subtract, multiple and divide two Pandas Series.  
#Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

s1=pd.Series([2, 4, 6, 8, 10])
s2=pd.Series([1, 3, 5, 7, 9])
print("Add")
print(s1.add(s2))
print("Subtract")
print(s1.sub(s2))
print("Multiply")
print(s1.mul(s2))
print("Divide")
print(s1.div(s2))
```

```python
#4. Write a Python program to get the powers of an array values element-wise.  

s1=pd.Series([0,1,2,3,4,5,6])
s2=pd.Series([3,3,3,3,3,3,3])
print("original array")
print(s1.values,s2.values)
res=s1.pow(s2)
print("First array elements raised to powers from second array, element-wise:  ")
print(res.values)
```

```python
#13. Write a Python program to select the rows where number of attempts in the 
#examination is less than 2 and score greater than 15.

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 
'Laura', 'Kevin', 'Jonas'], 
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], 
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']} 

df=pd.DataFrame(exam_data,index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])

print(df[(df["attempts"]<2) & (df["score"]>15)])

```
